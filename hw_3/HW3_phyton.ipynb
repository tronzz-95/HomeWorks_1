```python
df.groupby('Pclass')['Survived'].sum()/df['Pclass'].value_counts() #считаем процент выживших в каждом классе
```

```python
most_popular_male_name = df[df['Sex'] == 'male']['Name'].mode()[0]

most_popular_male_name
```

```python
most_popular_female_name = df[df['Sex'] == 'female']['Name'].mode()[0]

most_popular_female_name
```

```python
def most_popular_name(group):
    return group['Name'].mode()[0]
```

```python
df.groupby(['Pclass', 'Sex']).apply(most_popular_name) #Выводит нам информацию о популярных именах в каждом классе среди каждого пола

```

```python
#Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека)
df.groupby('Cabin').size() #считаем сколько раз номер кабины в списке вылез

```

```python
x = df.groupby('Cabin').size().value_counts().sort_index()  #подсчет количества 1,2,3,4 местных кабин

x
```

```python
for num_people, count in x.items():
    print(f"Количество кабин с {num_people} пассажирами: {count}")
```
